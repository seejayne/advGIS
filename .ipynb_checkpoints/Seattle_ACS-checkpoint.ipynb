```python
import pandas as pd
```

```python
df_sea = pd.read_csv('data/Census_Tracts_2010.csv')

# This changes the census tract column to be imported as a string
df_sea['GEOID10'] = df_sea['GEOID10'].astype(str)

df_sea
```

```python
df_acs_inc = pd.read_csv('data/acs2019_5yr_B19013_14000US53033002400.csv')

# This changes the census tract column to be imported as a string
df_acs_inc['GEOID10'] = df_acs_inc['GEOID10'].astype(str)

df_acs_inc = df_acs_inc[['GEOID10','B19013001']].copy()

df_acs_inc = df_acs_inc.rename(columns={"B19013001": "med_hh_inc"})
df_acs_inc
```

```python
df_acs_howner = pd.read_csv('data/acs2019_5yr_B25003_14000US53033002400.csv')

# This changes the census tract column to be imported as a string
df_acs_howner['GEOID10'] = df_acs_howner['GEOID10'].astype(str)
df_acs_howner = df_acs_howner[['GEOID10','B25003001','B25003002']].copy()
df_acs_howner= df_acs_howner.rename(columns={'B25003001': 'total','B25003002': 'own_occu'})
df_acs_howner
```

```python
from functools import reduce

sea_acs = [df_sea, df_acs_inc, df_acs_howner]
sea_acs_merged = reduce(lambda  left,right: pd.merge(left,right,on=['GEOID10'],
                                            how='outer'), sea_acs)
sea_acs_merged
```

```python
sea_acs_merged = sea_acs_merged[['GEOID10','med_hh_inc','total','own_occu']].copy()

sea_acs_merged
```
