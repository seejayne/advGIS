```python
import pandas as pd
```

```python
df_la = pd.read_csv('data/LA_County_Census_Tracts_2015.csv')
df_la
# This changes the census tract column to be imported as a string
df_la['FIPS'] = df_la['FIPS'].astype(str)

# And now we add the zero at the begining of the Census Tract Column
df_la['FIPS'] = df_la['FIPS'].str.zfill(11)

df_la= df_la.rename(columns={'FIPS': 'GEOID10'})

df_la
```

```python
df_acs_inc = pd.read_csv('data/acs2019_5yr_B19013_14000US06037433602.csv')

#This changes the census tract column to be imported as a string
df_acs_inc['GEOID10'] = df_acs_inc['GEOID10'].astype(str)

# And now we add the zero at the begining of the Census Tract Column
df_acs_inc['GEOID10'] = df_acs_inc['GEOID10'].str.zfill(11)

df_acs_inc = df_acs_inc[['GEOID10','B19013001']].copy()

df_acs_inc = df_acs_inc.rename(columns={"B19013001": "med_hh_inc"})
df_acs_inc
```

```python
df_acs_howner = pd.read_csv('data/acs2019_5yr_B25003_14000US06037433602.csv')

# This changes the census tract column to be imported as a string
df_acs_howner['GEOID10'] = df_acs_howner['GEOID10'].astype(str)

# And now we add the zero at the begining of the Census Tract Column
df_acs_howner['GEOID10'] = df_acs_howner['GEOID10'].str.zfill(11)

df_acs_howner = df_acs_howner[['GEOID10','B25003001','B25003002']].copy()
df_acs_howner= df_acs_howner.rename(columns={'B25003001': 'total','B25003002': 'own_occu'})
df_acs_howner
```

```python
from functools import reduce

la_acs = [df_la, df_acs_inc, df_acs_howner]
la_acs_merged = reduce(lambda  left,right: pd.merge(left,right,on=['GEOID10'],
                                            how='outer'), la_acs)
la_acs_merged
```

```python
la_acs_merged = la_acs_merged[['GEOID10','med_hh_inc','total','own_occu']].copy()

la_acs_merged.tail(20)
```

```python
la_acs_merged.drop(la_acs_merged.tail(5).index,inplace=True) # drop last n rows
la_acs_merged
```

```python
la_acs_merged.to_csv(r'la_acs_merged.csv')
```
